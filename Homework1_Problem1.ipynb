```python
# Importing relavent libraries... 
import torch                  # pytorch                
from numpy import asarray     # numpy asarry
from PIL import Image         # pillow Image 
```

Image1 analysis with Red object:

```python
print('Analysis for following image: ')
imgRed1 = Image.open('/content/drive/MyDrive/Photos/red1.jpeg')
imgRed1
```

```python
imgRed1_numpy = asarray(imgRed1)                                 # Converting the image to numpy array...
imgRed1_tensor = torch.from_numpy(imgRed1_numpy).float()         # Converting image to tensor of flaots...
Brightness = imgRed1_tensor.mean()                               # Brightness through mean of tensor..
R_value,G_value,B_value = imgRed1_tensor.mean(dim = [0,1])       # Getting average of each channel...
# Printing the results...
print('Image analysis for above image: ')
print('Brightness of Red Image 1: ',Brightness.item())
print('Red channel average: ',R_value.item())
print('Green channel average: ',G_value.item())
print('Blue channel average: ',B_value.item())
```





*   Brightness of Image1 is relatively moderate.
*   From image analysis, with red channel value, red item can be identified seamlessly.
